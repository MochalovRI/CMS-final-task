````python
import numpy as np
import scipy as sp
from scipy.optimize import brentq
````

````python
pv = 100000
fv = 0
````

````python
pmt1 = -1500
n1 = 120
````

````python
pmt2 = -1250
n2 = 180
````

````python
pmt3 = -1160
n3 = 240
````

````python
def fun1(i):
    return pv*(1+i)**n1 + pmt1*(((1+i)**n1-1)/i) + fv 
def fun2(i):
    return pv*(1+i)**n2 + pmt2*(((1+i)**n2-1)/i) + fv 
def fun3(i):
    return pv*(1+i)**n3 + pmt3*(((1+i)**n3-1)/i) + fv 
````

````python
sol1 = brentq(fun1, 0.01, 1.0)
sol1
````

````python
sol2 = brentq(fun2, 0.01, 1.0)
sol2
````

````python
sol3 = brentq(fun3, 0.01, 1.0)
sol3
````

````python
def solve_fun(pv, pmt, n, fv):
    def fun(i):
        return pv*(1+i)**n + pmt*((((1+i)**n)-1)/i) + fv
    return brentq(fun, 0.001, 1.0)
````

````python
i1 = solve_fun(pv, pmt1, n1, fv)
print(sol1)
i1
````

````python
i2= solve_fun(pv, pmt2, n2, fv)
print(sol2)
i2
````

````python
i3= solve_fun(pv, pmt3, n3, fv)
print(sol3)
i3
````

````python
iy1 = 12*i1
iy2 = 12*i2
iy3 = 12*i3
print("I1year=", iy1*100,'%')
print("I2year=", iy2*100,'%')
print("I3year=", iy3*100,'%')
````

# Отчёт

Функция, решающая уравнение $ PV*(1+i)^n + PMT*\frac{(1+i)^n-1}{i} + FV = 0 $ (1) выглядит следующим образом: 
```
def solve_fun(pv, pmt, n, fv):
    def fun(i):
        return pv*(1+i)**n + pmt*((((1+i)**n)-1)/i) + fv
    return brentq(fun, 0.001, 1.0)
```
Для решения уравнения (1) был исрользован метод Брента как наиболее популярный и универсальный.
